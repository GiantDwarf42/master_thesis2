```python
latex_df = df_b01.round(3)
```

```python
# Step 2: Flatten the MultiIndex to avoid \multirow issues
latex_header = [' & '.join([str(level) for level in header]) for header in latex_df.index.names]
latex_header = ' & '.join(latex_header) + ' \\\\ \n'

# Step 3: Convert DataFrame to LaTeX format with float_format
latex_code = latex_df.to_latex(multicolumn=False, multirow=False, float_format="{:.3f}".format)

# Step 4: Manually insert the custom header into the LaTeX code
latex_code = latex_code.replace(r'\toprule', r'\toprule' + latex_header)

# Option 1: Use \resizebox to fit the table to the text width
latex_code = (
    r"\begin{table}[htbp]" "\n"  # Start the table environment
    r"\centering" "\n"           # Center the table
    r"\resizebox{\textwidth}{!}{"  # Start the resizebox command
    + latex_code +                 # Add the LaTeX table code
    r"}" "\n"                       # Close the resizebox command
    r"\caption{Your Caption Here}" "\n"  # Add your caption here
    r"\label{tab:your_label_here}" "\n"  # Add a label for referencing
    r"\end{table}"                 # End the table environment
)

# Step 5: Define the folder and filename where you want to save the file

folder_path = 'G:/My Drive/Studium/UNIGE_Master/Thesis/Master_Thesis/Data/Gauss_server'
file_name = 'Gauss_0.1.txt'



# Step 6: Save the LaTeX code to the specified folder as a .txt file
file_path = os.path.join(folder_path, file_name)
with open(file_path, 'w') as file:
    file.write(latex_code)

print(f"LaTeX code saved to {file_path}")
```
